ANO: 2024

```python
# Define o ano diretamente
ano = 2024

# Verifica se o ano é bissexto
if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0):
    print(f"O ano {ano} é bissexto.")
else:
    print(f"O ano {ano} não é bissexto.")
```

10- Peça a quantidade de horas trabalhadas e o valor por hora. Se trabalhou até 40 horas, calcule o salário normalmente (horas × valor por hora). Se trabalhou mais de 40 horas, as horas extras devem ter um adicional de 50% (horas extras × valor por hora × 1.5). Crie um algoritmo em Python que realize esse cálculo corretamente.

HORAS DE TRABALHO: 45

VALOR POR HOEA: 20

```python
# Define as horas trabalhadas e o valor por hora
horas_trabalhadas = 45
valor_por_hora = 20

# Calcula o salário
if horas_trabalhadas <= 40:
    salario = horas_trabalhadas * valor_por_hora
else:
    horas_extras = horas_trabalhadas - 40
    salario = (40 * valor_por_hora) + (horas_extras * valor_por_hora * 1.5)

print(f"O salário total é: R$ {salario:.2f}")
```
